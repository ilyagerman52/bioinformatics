### Interesting mini-research.

The interesting thing is that the last mutation is actually removing some nucleobases. Maybe, this part of genome has no useful functions, so it can be removed by evolution process. (This is not an isolated case, so you can find the same case in [extra-screens](https://github.com/ilyagerman52/bioinformatics/tree/main/homework3/mutations/extra))

Another interesting thing is that the majority of mutation was in the place 25000+ in sequences.

I show another situation, because I thought that as closer mutations are, as much possible that they are in one gene. So, I tried to find mutation in different parts of gemone. In fact, the situation is opposite.

To explain that, the gene ORF1ab carries the main info about coronavirus. It matches big part of sequence, but it is more stable and more resistance to mutation.
