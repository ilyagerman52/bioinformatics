```python
print('main 4:\n')
!grep 'ENSG00000075035' simple_somatic_mutation.open.tsv | grep -o 'MU[0-9]*' | sort -u
print('-' * 100)
!grep 'ENSG00000132321' simple_somatic_mutation.open.tsv | grep -o 'MU[0-9]*' | sort -u
print('-' * 100)
!grep 'ENSG00000177103' simple_somatic_mutation.open.tsv | grep -o 'MU[0-9]*' | sort -u
print('-' * 100)
!grep 'ENSG00000251322' simple_somatic_mutation.open.tsv | grep -o 'MU[0-9]*' | sort -u
print('-' * 100)

print('\n\nextras:\n')
!grep 'ENSG00000114544' simple_somatic_mutation.open.tsv | grep -o 'MU[0-9]*' | sort -u
print('-' * 100)
!grep 'ENSG00000005238' simple_somatic_mutation.open.tsv | grep -o 'MU[0-9]*' | sort -u
```
